```python
%pip install matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data structure:[restaurant_id,2021,2022,2023,2024]
sales_data= np.array([
    [1, 150000, 180000,220000,250000], #Paradise Biryani
    [2,120000, 140000,160000, 190000], #Bejiing Bites
    [3,200000, 230000, 260000, 300000], #Pizza Hub
    [4, 180000, 210000, 240000, 270000], #Burger Point
    [5, 160000, 185000, 205000, 230000] # Chai Point
])

print("=============Zomato Sales Analysis=======================")
print("\n Sales Data Shape:", sales_data.shape)
```

```python
print("\n Sample Data for 1st 3 Restaurants:")
for i in range(3):
    print(f"Restaurant ID: {sales_data[i, 0]}, Sales: {sales_data[i, 1:]}")
```

```python
#total sales for all restaurant per year
print(np.sum(sales_data[:, 1:], axis=0))
#another method todo so the above method 
yearly_total=np.sum(sales_data[:, 1:], axis=0)
print("\nTotal Sales for all Restaurants per Year:", yearly_total)




#Minimum sales per restaurant 
min_sales = np.min(sales_data[:, 1:], axis=1)
print("\nMinimum Sales per Restaurant per Year:", min_sales)


# Maximum sales per restaurant
max_sales = np.max(sales_data[:, 1:], axis=0)
print("\nMaximum Sales per Restaurant per Year:", max_sales)


#Averwage sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("\nAverage Sales per Restaurant per Year:", avg_sales)
```

```python
cumsum= np.cumsum(sales_data[:, 1:], axis=1)
print("\nCumulative Sales per Restaurant per Year:",cumsum)
```

```python
plt.bar(sales_data[:, 0], min_sales, color='blue', label='Minimum Sales')
# Maximum sales per restaurant
max_sales = np.max(sales_data[:, 1:], axis=1)
plt.bar(sales_data[:, 0], max_sales, color='red', label='Maximum Sales', alpha=0.5)
# Average sales per restaurant
plt.bar(sales_data[:, 0], avg_sales, color='green', label='Average Sales', alpha=0.5)
plt.xlabel('Restaurant ID')
plt.ylabel('Sales')
plt.title('Sales Analysis of Restaurants')
plt.xticks(sales_data[:, 0], [f'Restaurant {int(id)}' for id in sales_data[:, 0]])
plt.legend()
plt.tight_layout()  
plt.show()
# Plotting cumulative sales
plt.figure(figsize=(10, 6))
plt.plot(sales_data[:, 0], cumsum[:, 0], marker='o', label='2021')  
plt.plot(sales_data[:, 0], cumsum[:, 1], marker='o', label='2022')
plt.plot(sales_data[:, 0], cumsum[:, 2], marker='o', label='2023')
plt.plot(sales_data[:, 0], cumsum[:, 3], marker='o', label='2024')
plt.xlabel('Restaurant ID')
plt.ylabel('Cumulative Sales')
plt.title('Cumulative Sales per Restaurant per Year')
plt.xticks(sales_data[:, 0], [f'Restaurant {int(id)}' for id in sales_data[:, 0]])
plt.legend()
plt.tight_layout()
plt.show()
```

```python
vector1= np.array([1, 2, 3, 4, 5])
vector2= np.array([6, 7, 8, 9, 10])

print("\nVector addition:", vector1 + vector2)
print("Vector subtraction:", vector1 - vector2)
print("Vector multiplication:", vector1 * vector2)
print("Vector division:", vector1 / vector2)
print("Vector dot product:", np.dot(vector1, vector2))

angle= np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Angle between vectors in radians:", angle)
```

```python
restaurant_types = np.array(['Biryani', 'Chinese', 'Pizza', 'Burger', 'Cafe'])
vectorised_upper = np.vectorize(str.upper)
print("\nRestaurant Types in Uppercase:", vectorised_upper(restaurant_types))
# Vectorized operations on restaurant types
vectorised_length = np.vectorize(len)
print("Length of each restaurant type:", vectorised_length(restaurant_types))
```
